# Problema de negocio
Se trata de un algoritmo capaz de analizar las caracteristicas de una propiedad como las comodidades,tamaño, ocupación del espacio en un periodo determinado y sugerir al anfitrion cual es  por precios de a cobrar para generar ingresos en ciertas temporadas de alta demanda

## Entendiendo el problema


```python
import pandas as pd
#inportando datos
ruta_acceso = '/content/drive/MyDrive/Pandas_Transformacion_de_datos/2071-aula1/data/datos_hosting.json'
datos = pd.read_json(ruta_acceso)
datos.head()
```

```python
#extrayendo detalles de las columnas
datos = pd.json_normalize(datos['info_inmuebles'])
datos.head()
```

```python
datos.info
```

## 2. Datos numéricos


```python
#veremos cuales columnas tiene nuestro conjunto de datos
columnas  = list(datos.columns)
columnas
```

```python
# desde la posición 3 en adelante haremos explode
datos = datos.explode(columnas[3:])
```

```python
#ya tengo los registros separados
datos
```

```python
#reseteando los indices
datos = datos.reset_index(drop=True)
datos.head()
```

```python
datos.info()
```

```python
# asignando tipos de datos correctos al conjunto
 #0   evaluacion_general    3818 non-null   object
# 1   experiencia_local     3818 non-null   object
# 2   max_hospedes          3818 non-null   object
# 3   descripcion_local     3818 non-null   object
# 4   descripcion_vecindad  3818 non-null   object
 #5   cantidad_baños        3818 non-null   object
 #6   cantidad_cuartos      3818 non-null   object
 #7   cantidad_camas        3818 non-null   object
# 8   modelo_cama           3818 non-null   object
# 9   comodidades           3818 non-null   object
# 10  cuota_deposito        3818 non-null   object
 #11  cuota_limpieza        3818 non-null   object
# 12  precio                3818 non-null   object
import numpy as np
```

```python
columnas_numericas = ['max_hospedes','cantidad_baños','cantidad_cuartos','cantidad_camas']
datos[columnas_numericas] = datos[columnas_numericas].astype(np.int64)
```

```python
datos['evaluacion_general'] = datos['evaluacion_general'].astype(np.float64)
```

```python
datos['precio'] = datos['precio'].apply(lambda x: x.replace('$','').replace(',','').strip())
datos['precio'] = datos['precio'].astype(np.float64)
```

```python
columnas_monetarias = ['cuota_deposito','cuota_limpieza']
datos[columnas_monetarias] = datos[columnas_monetarias].apply(lambda x: x.str.replace('$','').str.replace(',','').astype(np.float64))
```

```python
#mostrando columnas monetarias
datos[['cuota_deposito','cuota_limpieza','precio']]
```

```python
datos.info()
```

## 3. Datos de texto

```python
datos.head()
```

empezaremos a toquenizar para la descripcion del local

```python
datos['descripcion_local']
```

```python
# haciendo columna como string
datos['descripcion_local'] = datos['descripcion_local'].str.lower()
```

```python
datos.head()
```

```python
datos.info()
```

```python
datos['descripcion_local'][3169]
```

```python
datos['descripcion_local'] = datos['descripcion_local'].str.replace('[^a-zA-Z0-9\-\']',' ',regex=True)
```

```python
datos['descripcion_local'] = datos['descripcion_local'].str.replace('(?<!\w)-(?!\w)',' ',regex=True)
```

```python
datos['descripcion_local'].head()
```

### Tokenización

```python
#separar cada una de las frases por medio de palabras por el espacio
datos['descripcion_local'] = datos['descripcion_local'].str.split()
```

```python
datos.info()
```

```python
datos['comodidades'] = datos['comodidades'].str.replace('\{|}|\"','', regex=True)
```

```python
datos['comodidades'] = datos['comodidades'].str.split(',')
datos.head()
```
